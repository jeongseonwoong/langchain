2. 프롬프트에 변수를 사용하기 위해서는 프롬프트 템플릿 안에 튜플 리스트를 넣어서 사용

PromptTemplate
단순 텍스트 템플릿
role 구분 없음 (system, user, assistant 없음)
단순한 질문/답변에 적합

ChatPromptTemplate
role 구분 가능 (system, user, assistant)
대화형 모델에 최적화
system 프롬프트 넣을 수 있음

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate(
    template="""
    What is the Capital of {country}?
    """,
    input_variables=["country"],
)

prompt = prompt_template.invoke({"country": "France"})

print(prompt)

llm.invoke(prompt)
```

```python
from langchain_core.prompts import ChatPromptTemplate
chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that can answer questions about the world."),
    ("user", "What is the capital of {country}?"),
])

chat_prompt = chat_prompt_template.invoke({"country": "France"})

print(prompt)

llm.invoke(chat_prompt)
```

3. 프롬프트에 변수를 사용할 필요가 없을 때는 템플릿 없이 튜플 리스트 사용

```python
from langchain_core.messages import HumanMessage, SystemMessage

message_list = [
    ("system", "You are a helpful assistant that can answer questions about the world."),
    ("user", "What is the capital of France?"),
]
llm.invoke(message_list)
```

```python
messages = [    
    ("system", """
    너는 영어회화를 연습하기 위한 상대야.
    아래 예시처럼 답해:
    User: What do you do for a living?
    Assistant: I'm a software engineer at a tech company. I mostly build backend services.

    User: Do you have any hobbies?
    Assistant: Yeah, I love hiking on wekends. It helps me clear my mind.

    User: What is your favorite food?
    Assistant: I like pizza and sushi. But I try to eat healthy most of the time.

    User: What movie did you watched last week?
    Assistant: I watched "The Dark Knight" last week. It's a great movie.

    User: What is your favorite color?
    Assistant: I like blue. It's calming and reminds me of the sky.

    User: What is your favorite season?
    Assistant: I like spring. The flowers are beautiful and the weather is nice.

    User: What is your favorite holiday?
    Assistant: I love Christmas. It's a time for family and friends.
    """),
    ]


while True:
    user_input = input("You: ")
    messages.append(("user", user_input))

    reply = llm.invoke(messages)
    print(f"Assistant: {reply.content}")
    messages.append(("assistant", reply.content))
```
