```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="qwen2.5:7b",
                 think = False,
                 temperature = 0)
                
llm
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

food_template = PromptTemplate(
    template = """
    What is the favorite food from {country}.
    Return only the name of the food
    """,
    input_variables=["country"]
)

food_chain = food_template | llm |StrOutputParser()

```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

recipe_template = ChatPromptTemplate.from_messages([
    ("system",
    """
    Provide the recipe of the food that the user wants
    respond only in Korean. Never use any other country's language.
    Response must except XML tags. only include content
    <examples>
        <example>
        1. 재료: 양파, 감자, 닭도리탕 용 닭고기, 고추가루, 고추장, 물, 다진 마늘, 간장
        2. 닭도리탕 용 닭을 끓는 물에 살짝 데친 후 찬물에 씻어 핏기를 뺍니다.
        3. 씻은 닭을 프라이팬에 놓고 간장 반컵컵, 고춧가루 3스푼, 다진마늘 2T스푼을 넣고 섞어줍니다.
        ...
        </example>
    </examples>
    """),
    ("human", "Can you give me the recipe for making {food}")
    ])

recipe_chain = food_chain | recipe_template | llm | StrOutputParser()

answer = recipe_chain.invoke({"country" : "USA"})

answer
```
